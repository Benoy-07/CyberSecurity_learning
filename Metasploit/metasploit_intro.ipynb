# 🛠️ Metasploit 7 Module Types

## 1️⃣ Exploit

**কাজ:**  
টার্গেট সিস্টেমের কোনো vulnerability ব্যবহার করে ঢোকার চেষ্টা করা।

**সহজ উদাহরণ:**  
ধরা যাক, একটা পুরনো Windows XP মেশিনে SMB vulnerability আছে।  
👉 Exploit module সেই দুর্বলতা ব্যবহার করে সিস্টেমে access নেওয়ার চেষ্টা করবে।

**🧠 বাস্তব ধারণা:**  
> “দরজার লক ভাঙার টুল”

---

## 2️⃣ Payload

**কাজ:**  
Exploit সফল হলে, টার্গেট মেশিনে কি কাজ হবে সেটা নির্ধারণ করে।

**সহজ উদাহরণ:**  
- Reverse shell (টার্গেট থেকে attacker-এর কাছে কানেকশন যাবে)  
- Meterpreter session (advanced control)

**🧠 বাস্তব ধারণা:**  
> “দরজা খুলে ঢুকে ভিতরে কী করবে”

---

## 3️⃣ Auxiliary

**কাজ:**  
Exploit না করেও scanning, enumeration, brute-force ইত্যাদি কাজ করা।

**সহজ উদাহরণ:**  
- কোন পোর্ট খোলা আছে তা দেখা  
- FTP login brute-force করা  
- HTTP server version চেক করা

**🧠 বাস্তব ধারণা:**  
> “আগে রেকি করা (information gathering)”

---

## 4️⃣ Post

**কাজ:**  
একবার সিস্টেমে ঢুকে গেলে, ভিতরের তথ্য সংগ্রহ বা কাজ করা।

**সহজ উদাহরণ:**  
- Password hash বের করা  
- Logged-in user দেখা  
- Screenshot নেওয়া

**🧠 বাস্তব ধারণা:**  
> “ঘরের ভেতরে ঢুকে আলমারি খোঁজা”

---

## 5️⃣ Encoder

**কাজ:**  
Payload-কে encode করা যেন antivirus সহজে ধরতে না পারে।

**সহজ উদাহরণ:**  
- Payload-এর byte pattern বদলে দেওয়া  
- Same কাজ কিন্তু ভিন্ন signature

**🧠 বাস্তব ধারণা:**  
> “চেহারা বদলে চলাফেরা করা”

---

## 6️⃣ NOP

**কাজ:**  
Exploit stability ঠিক রাখা, buffer overflow-এ সাহায্য করা।

**সহজ উদাহরণ:**  
- CPU-কে কিছু না করার instruction দিয়ে timing ঠিক রাখা

**🧠 বাস্তব ধারণা:**  
> “ফাঁকা সময় পূরণ করা”

---

## 7️⃣ Evasion

**কাজ:**  
Modern antivirus, IDS/IPS, defender এড়িয়ে চলা।

**সহজ উদাহরণ:**  
- Sandbox detection bypass  
- Obfuscation technique

**🧠 বাস্তব ধারণা:**  
> “CCTV এড়িয়ে চুপচাপ চলা”


#### In sort
- Exploit   → ঢোকা
- Payload   → কী করবে
- Auxiliary → খোঁজাখুঁজি
- Post      → ঢোকার পর কাজ
- Encoder   → লুকানো
- NOP       → Stability
- Evasion   → Security bypass
